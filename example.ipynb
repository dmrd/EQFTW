# Triangle with premade cuts -> Square

```python
# Triangle w/ cuts -> rectangle -> square w/ debugging lines
tri = e.random_triangle()
for i in range(5):
    tri = e.random_cut_shape(tri)
tri.plot();e.show()

r = e.tri2rect(tri, plot=True)
r.plot();e.show()
s = e.rect2square(r, plot=True)
s.plot()
_, _, hx, hy = s.bbox()
o = s.original_position()
e.axis_align(o)
o.plot((hx + 5, 0))
e.show()
```

```python
# Uncut triangle -> square
tri = e.random_triangle()
r = e.tri2rect(tri)
s = e.rect2square(r)
_, _, hx, hy = s.bbox()
s.plot()
o = s.original_position()
e.axis_align(o)
o.plot((hx + 5, 0))
e.show()
```
